Приведем данные Ильсеяр к нормальному виду

```python
import json
data = json.load(open('/data2/polyakov/instruction_tuning/ToolBench/data/toolbench_multistep_dataset_g_query_format_100.json', 'r'))
```

```python
new_names = {
    'Measurement Units Converter': 'measurement_units_converter',
    'Convert from one unit of measure to another': 'm_one_unit_of_measure_to_another',
    'Convexity': 'convexity',
    'hsl to rgb': 'hsl_to_rgb'
}
```

```python
for item in data:
    relevant_apis = item['relevant APIs']
    new_relevant_apis = []
    for api in relevant_apis:
        new_apis = []
        for name in api:
            if name in new_names:
                new_apis.append(new_names[name])
            else:
                new_apis.append(name)
        new_relevant_apis.append(new_apis)
    item['relevant APIs'] = new_relevant_apis
```

```python
for item in data:
    print(item['relevant APIs'])
```
