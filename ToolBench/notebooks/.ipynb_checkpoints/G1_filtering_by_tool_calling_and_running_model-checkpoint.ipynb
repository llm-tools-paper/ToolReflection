Отфильтруем тулы по вызываемости

```python
all_callable_examples = []
for item in tqdm(data[:5]):
    flag = True
    for step in item['conversations']:
        if step['from'] == 'assistant':
            tool_name, tool_input = step['value'].split('Action: ')[1].split('\nAction Input: ')
            tool_output = call_tool(tool=tool_name, tool_input=tool_input, 
                                    toolbench_key=toolbench_key, 
                                    tool_to_url_data_path=tool_to_url_data_path)
            if tool_name != 'Finish':
                if tool_output.split('"error":"')[1].split('",')[0]:
                    flag = False
    if flag:
        all_callable_examples.append(item)   
```

```python
with open('data/self_correction/train_data_cleaned_g1_callable.json', 'w') as f:
    json.dump(all_callable_examples, f, ensure_ascii=False)
```

Составим промпт ровно таким же образом, как это делается на инференсе

```python
def make_prediction(backbone_model, messages):
    conv = get_conversation_template(backbone_model.template)
    roles = {"system": conv.roles[0], "user": conv.roles[1], "function": conv.roles[2], "assistant": conv.roles[3]}
    prompt = ''
    for message in messages:
        role = roles[message['from']]
        content = message['value']
        prompt += f"{role}: {content}\n"
    prompt += "Assistant:\n"
    print(prompt)
    prediction = backbone_model.prediction(prompt)
    return prediction
```
