```python
import json
```

G1

```python
with open('data/instruction/G1_query.json', 'r') as f:
    queries = json.load(f)
```

```python
with open('data/test_query_ids/G1_instruction_test_query_ids.json', 'r') as f:
    inst_ids = json.load(f)
with open('data/test_query_ids/G1_category_test_query_ids.json', 'r') as f:
    cat_ids = json.load(f)
with open('data/test_query_ids/G1_tool_test_query_ids.json', 'r') as f:
    tool_ids = json.load(f)
```

```python
inst_examples = []
cat_examples = []
tool_examples = []

for item in inst_ids.keys():
    for example in queries:
        if example['query_id'] == int(item):
            inst_examples.append(example)
for item in cat_ids.keys():
    for example in queries:
        if example['query_id'] == int(item):
            cat_examples.append(example)
for item in tool_ids.keys():
    for example in queries:
        if example['query_id'] == int(item):
            tool_examples.append(example)
```

```python
inst_examples
```

```python
with open('data/test_data_merged/G1_inst.json', 'w') as f:
    json.dump(inst_examples, f, ensure_ascii=False)
with open('data/test_data_merged/G1_cat.json', 'w') as f:
    json.dump(cat_examples, f, ensure_ascii=False)
with open('data/test_data_merged/G1_tool.json', 'w') as f:
    json.dump(tool_examples, f, ensure_ascii=False)
```
