```python
import os
from pathlib import Path
import json
```

```python
def print_errors(dir):
    print('Length: ', len(os.listdir(dir)))
    count = 0
    file_count = 0
    num_calls = 0
    for file in os.listdir(dir):
        flag = False
        with open(dir+file, 'r') as f:
            curr_ans = json.load(f)
        for step in curr_ans['trys'][0]['chain']:
            if step['node_type'] == 'Action Input':
                num_calls += 1
                if step['observation_code'] == 1 or step['observation_code'] == 2:
                    print(step)
                    print('#' * 100)
                    count += 1
                    flag = True
                else:
                    observation = step['observation']
                    try:
                        observation = observation.split('\"error\": \"')[1].split('\", \"response\"')[0]
                        if observation:
                            if 'unexpected' in observation or 'missing' in observation:
                                print(observation)
                                print(step)
                                print('#' * 100)
                                print(file)
                                flag = True
                                count += 1
                    except:
                        pass
        if flag:
            file_count += 1
    print('Error file count:', file_count)
    print('Num calls:', num_calls)
    print('Count:', count)
    print('Error rate:', count / num_calls)
```

```python
print_errors('data/answer_toollama/toollama2/G1_inst/')
```
